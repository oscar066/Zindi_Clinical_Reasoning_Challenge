```python
import pandas as pd

# Load the dataset
file_path = "/Users/naomiamadi/Downloads/Documents/Zindi Clinical Reasoning Challenge/train.csv"
df = pd.read_csv(file_path)

# Preview the first few rows
print(df.head())
```

```python
# Convert column names to lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# List of columns to clean
text_columns = ['county', 'health_level', 'prompt', 'nursing_competency',
                'clinical_panel', 'clinician', 'gpt4.0', 'llama', 'gemini', 'ddx_snomed']

# Strip leading/trailing spaces and convert to lowercase
df[text_columns] = df[text_columns].applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)
```

```python
# Convert to numeric and coerce errors (e.g., empty strings become NaN)
df['years_of_experience'] = pd.to_numeric(df['years_of_experience'], errors='coerce')
```

```python
# Drop duplicates based on key columns
df.drop_duplicates(subset=['master_index', 'prompt', 'clinician'], inplace=True)
```

```python
# Drop rows with missing essential data (e.g., model outputs)
df.dropna(subset=['prompt', 'clinician', 'gpt4.0', 'llama', 'gemini', 'ddx_snomed'], inplace=True)
```

```python
# Filter out unrealistic experience values (e.g., > 60 years)
df = df[(df['years_of_experience'] >= 0) & (df['years_of_experience'] <= 60)]
```
