### Data Preparation and Tokenization

```python
from datasets import Dataset
```

```python
# Convert your pandas DataFrames to Hugging Face Dataset objects

train_dataset = Dataset.from_pandas(train_data)

test_dataset = Dataset.from_pandas(test_data)
```

```python
train_dataset = train_dataset.train_test_split(test_size=0.2)
print(train_dataset)
```

```python
from transformers import AutoTokenizer, DataCollatorForSeq2Seq

# Use the tokenizer for the model you plan to fine-tune
model_name = "t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def preprocess_function(examples):
    # T5 requires a prefix for different tasks. We can create one.
    prefix = "diagnose the following clinical case: "
    inputs = [prefix + doc for doc in examples['prompt']]
    
    # The 'text_target' is the 'clinician' column
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True, padding="max_length")
    
    # Tokenize the targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples['clinician'], max_length=512, truncation=True, padding="max_length")
        
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# # Apply the function to your entire dataset
# tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)

# tokenized_test_dataset = test_dataset.map(preprocess_function, batched=True)
```

```python
# preprocess with the other columns as context
def preprocess_with_context(examples):
    # The target text
    targets = examples['clinician']
    
    # Build the structured input string for each example
    inputs = []
    for i in range(len(targets)):
        # Construct the context string
        context = f"context: county [{examples['county'][i]}] health_level [{examples['health_level'][i]}] experience [{examples['years_of_experience'][i]}]"
        
        # Combine with the prompt
        prompt = examples['prompt'][i]
        
        # Final input string
        input_string = f"{context} prompt: {prompt}"
        inputs.append(input_string)
        
    # Tokenize the structured inputs
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True, padding="max_length")

    # Tokenize the targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length")
        
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# Apply the function to each dataset
tokenized_train_dataset = train_dataset["train"].map(preprocess_with_context, batched=True)

tokenized_test_dataset = train_dataset["test"].map(preprocess_with_context, batched=True)
```

```python
train_dataset.column_names
```

```python
# Remove all original columns, as they are now encoded in the 'input_ids' and 'labels'
original_columns = train_dataset["train"].column_names

tokenized_train_dataset = tokenized_train_dataset.remove_columns(original_columns)

tokenized_test_dataset = tokenized_test_dataset.remove_columns(original_columns)
```

```python
tokenized_train_dataset.column_names
```

```python
from transformers import T5ForConditionalGeneration, Trainer, TrainingArguments

# Load the pre-trained model
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./clinical_t5_model",      # Where to save the model
    num_train_epochs=3,                    # Number of times to go through the data
    per_device_train_batch_size=4,         # Batch size for training
    per_device_eval_batch_size=4,          # Batch size for evaluation
    warmup_steps=500,                      # Number of warmup steps for learning rate scheduler
    weight_decay=0.01,                     # Strength of weight decay
    logging_dir='./logs',                  # Directory for storing logs
    logging_steps=10,
    eval_strategy="epoch",           # Evaluate at the end of each epoch
    save_strategy="epoch",                 # Save the model at the end of each epoch
    load_best_model_at_end=True,           # Load the best model found during training
)
```

```python
# Create the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_test_dataset,
    tokenizer=tokenizer,
)

# Start fine-tuning
trainer.train()

# Save your fine-tuned model
trainer.save_model("./final_clinical_t5_model")
tokenizer.save_pretrained("./final_clinical_t5_model")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer, AutoTokenizer

# Load your fine-tuned model and tokenizer
model_path = "./final_clinical_t5_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)

# Example new prompt
new_prompt = "i am a nurse with 10 years of experience in general nursing working in a county referral hospital in nairobi county in kenya a 55 year old female presents with a 3 day history of shortness of breath and chest pain vital signs are unstable bp 90 50 pr 120 t 38.1 rr 28 spo2 88 question what are the immediate steps"

# Prepare the input for the model
prefix = "diagnose the following clinical case: "
input_ids = tokenizer(prefix + new_prompt, return_tensors="pt").input_ids

# Generate the output
output_ids = model.generate(input_ids, max_length=512, num_beams=4, early_stopping=True)
generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print(generated_text)
```
